# Definici칩n de funciones

```python
%display latex

#Definici칩n de variables

var('x y z t')                            #Coordenadas y tiempo
a = var(r'alpha')                         #Par치metro de deformaci칩n

#Densidad en equilibrio

n_eq= 1-x^2-y^2-(sqrt(8))^2*z^2

nxz(x,z)=n_eq.substitute(y==0)            #Plano xz
nyz(y,z)=n_eq.substitute(x==0)            #Plano yz
nxy(x,y)=n_eq.substitute(z==0)            #Plano xy

#Modos dipolares

n_111 = sqrt(3/pi)/2*x                    #l=1, m=1
n_112 = sqrt(3/pi)/2*y                    #l=1, m=-1
n_10 = sqrt(3/pi)/2*z                     #l=1, m=0

#Modos cuadrupolares

n_221 = sqrt(15/pi)/4*(x^2-y^2)           #l=2, m=2
n_222 = sqrt(15/pi)/2*x*y                 #l=2, m=-2
n_211 = sqrt(15/pi)/2*x*z                 #l=2, m=1
n_212 = sqrt(15/pi)/2*y*z                 #l=2, m=-1
n_20 = -(3/2*a^2+1/2*sqrt(9*a^4-16*a^2+16))/(2+3/2*a^2+1/2*sqrt(9*a^4-16*a^2+16))+x^2+y^2+(-2+3/2*a^2+1/2*sqrt(9*a^4-16*a^2+16))*z^2
n_00 = -(3/2*a^2-1/2*sqrt(9*a^4-16*a^2+16))/(2+3/2*a^2-1/2*sqrt(9*a^4-16*a^2+16))+x^2+y^2+(-2+3/2*a^2-1/2*sqrt(9*a^4-16*a^2+16))*z^2

n_000 = n_00.substitute(a==sqrt(8))       #l=0, m=0
n_200 = n_20.substitute(a==sqrt(8))       #l=2, m=0
```

# Densidad en equilibrio

```python
def Nxz(x,z): 
    if nxz(x,z)>=0:
        return nxz(x,z)
    else: 
        return 0
    
def Nyz(y,z): 
    if nyz(y,z)>=0:
        return nyz(y,z)
    else: 
        return 0
    
def Nxy(x,y): 
    if nxy(x,y)>=0:
        return nxy(x,y)
    else: 
        return 0
```

```python
contour_plot(Nxz, (x,-2,2),(z,-0.5,0.5),plot_points=200, cmap = 'jet')
```

```python
contour_plot(Nyz, (y,-2,2),(z,-0.5,0.5),plot_points=200, cmap = 'jet')
```

```python
contour_plot(Nxy, (x,-2,2),(y,-2,2),plot_points=200, cmap = 'jet')
```
